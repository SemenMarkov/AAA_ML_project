__1) Оценить потенциал проекта. Насколько важно решить задачу?__

- размытие фона помогает клиенту сразу увидеть товар, не отвлекаясь на другие предметы, что улучшает опыт покупателя и в перспективе может увеличивать конверсию
- размытие фона помогает убирать персональные данные и сторонние контакты, которые могли случайно попасть на фото, что помогает бороться с мошенниками
- обучение модели для размытия фона может быть полезно для обучения других моделей, например, модели распознавания товара на фото

__2) Есть ли простое решение? Насколько оно решит задачу? Сложно ли поддерживать такое решение?__

**Crop**

- Вариант простого решения:
    - берем для превью центральный фрагмент фото
        - Плюсы: 
            - очень простое решение задачи без использования ML
            - не требует много ресурсов
        - Минусы: 
            - опирается на предположение о том, что товар всегда находится в центре фото: то есть если товар на фото далеко, на preview он будет маленького размера; а если он сфотографирован крупным планом или не по центру, то на preview в заданную рамку полностью не будет помещаться;            
- С точки зрения поддержки: фото будут обрабатываться по одному и тому же скрипту

**Blur**

- Вариант простого решения:
    - по аналогии с Crop: размываем края фото, центр оставляем неразмытым 

__3) Реалистичность решения проблемы с помощью машинного обучения__

- такая задача уже решена во многих компаниях!
    - например в Zoom и Google Meet есть размытие фона в режиме реального времени
- по сути данная задача - типичная задача сегментации; нам необходимо просто подстроить существующую архитектуру под специфику Авито и дообучить модель на данных с Авито

__4) Особенности задачи__

- мы решаем задачу instance segmentation - то есть нам нужно найти границы всех предметов на фото
- затем, нам нужно принять решение о том, какие предметы относятся к товару, а какие просто попали в кадр: здесь можно воспользоваться разметкой на Толоке и эвристиками
- важно: мы не может ограничиться bounding boxes для товаров, т.к. размывать нужно именно по границе объекта => нужны подходящие метрики для измерения качества модели
