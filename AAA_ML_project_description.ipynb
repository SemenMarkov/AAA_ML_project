__1) Оценить потенциал проекта. Насколько важно решить задачу?__

- размытие фона заставляет клиента сразу видеть объект продажи, не отвлекаясь на другие объекты, что улучшает опыт покупателя и в перспективе может увеличивать конверсию
- размытие фона помогает убирать персональные данные и сторонние контакты, что минимизирует риск ухода клиентов с платформы до завершения сделки
- обучение модели для облегчения обучения и увеличения скора на других задачах посредством улучшения качества данных

__2) Есть ли простое решение? Насколько оно решит задачу? Сложно ли поддерживать такое решение?__

**Crop**

- Вариант простого решения:
    - вместо задачи сегментации, можно решать задачу детекции: находим объект продажи, обрезаем фото вне рамки объекта продажи.
        - Плюсы: 
            - доступность данных для обучения модели (Image Net, ...)
            - менее сложные архитектуры модели
        - Минусы: 
            - теряется масштаб объекта на фото
            - фон все равно необходимо размывать
- Данный вариант решает поставленную задачу частично, так как все еще остается фон в рамках границ изображения объекта продажи
- Сложность поддержки решения такая же как и у сегментации

- Вариант простого решения:
    - центрируем фото в рамке заданного размера
        - Плюсы: 
            - решение задачи без использования ML архитектур
        - Минусы: 
            - опирается на предположение того, что объект продажи всегда находится в центре фотографии: то есть если объект продажи на фото далеко, на preview он будет изображен полностью, но будет очень маленького размера; если объект продажи на фото находится очень крупно, то на preview в заданную рамку полностью не будет помещаться;            
- С точки зрения поддержки фото будут обрабатываться по одному и тому же простому алгоритму
